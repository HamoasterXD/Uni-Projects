2 - Transformation

a) Min Max Scaling:

```python
# min max equation as a function
def minmax(column, newmin, newmax):
    return (   (column - column.min())  / (column.max() - column.min() )    ) * ( newmax - newmin ) + newmin 
```

```python
# in this example i will take the "MetaScore" in my min max transformation

df_minmax = minmax(df.Metascore, 0, 1)
print(df_minmax)
```

b) Z- Scaling


```python
# same idea like the one above. we will define zscore equation as a function

def zscore(column):
    return ( column - column.mean() )   /   ( column.std() )
```

```python
# same as the above, i will apply zscore equation on "Metascore" 

df_zscore = zscore(df.Metascore)
print(df_zscore)
```

3) Clustering

a) partition clustering using kmeans 

```python
# create kmeans model ;(

import numpy as np

# define function to pick centroids randomly
def initialize_centroids(X, k):
    np.random.seed(60)  # can be any value. this is the random generator
    random_indices = np.random.permutation(X.shape[0]) # pick random value from X which is the column that we will apply k means on
    centroids = X[random_indices[:k]] # pick the values from 1 to number of clusters RANDOMLY
    return centroids

# this function is responsile for finding the nearest point
def assign_clusters(X, centroids): 
    distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2)) # computes distance between each point
    cluster_assignment = np.argmin(distances, axis=0) # selects shortest distance
    return cluster_assignment 

#  this function computes the new centroids based on the mean of the data points assigned to each cluster
def compute_centroids(X, labels, k):
    centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)]) # calculates mean after finding each shortest point. then it will take the mean point and set it as the main point for the next iteration
    return centroids

# this function is responsible for the iterations. i set max iterations to 60 as it is more than enough
def kmeans(X, k, max_iters=60, tol=1e-4):
    centroids = initialize_centroids(X, k)
    for i in range(max_iters):
        old_centroids = centroids # takes the mean value as the new center
        labels = assign_clusters(X, centroids) # assigns clusters for each data point 
        centroids = compute_centroids(X, labels, k) # assigns final centroids
        if np.all(np.abs(centroids - old_centroids) < tol):
            break
    return centroids, labels
```

```python
# yep you guessed it. we will apply the kmeans on Metascore

X = df[['Metascore']].values
k = 2
centroids, labels = kmeans(X, k)

df['Clusters'] = labels
```

b) Kmedioide

```python


class KMedoids:
    def __init__(self, n_clusters=2, max_iters=60):
        self.n_clusters = n_clusters
        self.max_iters = max_iters


    # define function to pick medioides randomly
    def initialize_medoids(self, X):
        np.random.seed(60) # can be any value. this is the random generator
        random_indices = np.random.permutation(X.shape[0]) # pick random value from X which is the column that we will apply k medioide on
        medoids = X[random_indices[:self.n_clusters]] # pick the values from 1 to number of clusters RANDOMLY
        return medoids

    # this function is responsile for finding the nearest point
    def assign_clusters(self, X, medoids): 
        distances = np.abs(X[:, np.newaxis] - medoids) # computes distance between each medioide
        cluster_assignment = np.argmin(distances.sum(axis=2), axis=1) # selects shortest distance
        return cluster_assignment 

    #  this function computes the new medioides based on the median of the data points assigned to each cluster
    def compute_medoids(self, X, labels): 
        medoids = np.zeros((self.n_clusters, X.shape[1])) 
        for i in range(self.n_clusters):
            cluster_points = X[labels == i] # select data points belonging to the current cluster
            distances = np.abs(cluster_points[:, np.newaxis] - cluster_points) # compute distance 
            total_distance = distances.sum(axis=2) # compute distance from each data point with another
            new_medoid_index = np.argmin(total_distance.sum(axis=1)) # finds the index of the data point with the minimum total distance
            medoids[i] = cluster_points[new_medoid_index] # update new medoide 
        return medoids

    # this function is responsible for the iterations
    def fit(self, X):
        medoids = self.initialize_medoids(X)
        for _ in range(self.max_iters):
            old_medoids = medoids # takes the old mediode value as the new center
            labels = self.assign_clusters(X, medoids) # assigns clusters for each data point 
            medoids = self.compute_medoids(X, labels) # assigns final centroids
            if np.allclose(np.abs(medoids - old_medoids), 0):
                break
            self.medoids = medoids
            return self

    # this function predicts each X values belongs to which cluster
    def predict(self, X):
        distances = np.abs(X[:, np.newaxis] - self.medoids) # calculates distance between X and the value of the medioide value 
        return np.argmin(distances.sum(axis=2), axis=1)
```

```python
# apply the k medioide

k = 2
kmedoids = KMedoids(n_clusters=k, max_iters=60)
kmedoids.fit(X)
```

```python
# extract clusters

labels_Medioide = kmedoids.predict(X)
df['ClusterMedioide'] = labels_Medioide
```

c) Hirarchial Clustering

```python
# import model 
from sklearn.cluster import AgglomerativeClustering

X = df.Metascore.head(5).values.reshape(-1, 1) # subsample only first 5 so we can plot the dendogram easily

agg_cluster = AgglomerativeClustering(n_clusters = 5, metric = 'euclidean', linkage = 'single')
agg_cluster.fit(X)
```

```python
# to draw a dendogram we need to create a linkage matrix
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(X)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15))
plt.xlabel('Index')
plt.ylabel('Distance')
dendrogram(Z, labels=df.index[:5].tolist())
plt.show()
```

d) Density Based Clustering

```python
# import model

from sklearn.cluster import DBSCAN

X = df['Duration (min)'].values.reshape(-1, 1)

dbscan = DBSCAN(eps=5, min_samples=10)  # eps is the maximum distance between two samples for them to be considered as in the same neighborhood and min_samples is the number of samples in a neighborhood for a point to be considered a core point.
dbscan.fit(X)
```

```python
# display
labels = dbscan.labels_
df['ClustersDBC'] = labels
```
