### Load Files

```python
import pandas as pd
import os

# Load dataset (modify the path if needed)
# Read all .txt files in the "books" directory
books_dir = "books"
files = [f for f in os.listdir(books_dir) if f.endswith('.txt')]

# Combine all text files into a single DataFrame
data = []
for file in files:
    with open(os.path.join(books_dir, file), 'r', encoding='utf-8') as f:
        text = f.read()
        data.append({"docno": file, "text": text})

df = pd.DataFrame(data)

# Required format: columns "docno" and "text"
df = df.rename(columns={"YourDocIDColumn": "docno", "YourTextColumn": "text"})
df = df[["docno", "text"]]
```

```python
# Preview the dataset
df.head()
```

### Indexing

```python
# Indexing
indexer = pt.IterDictIndexer("./shakespeare_index")
index_ref = indexer.index(df.to_dict(orient="records"))

# Load the index
index = pt.IndexFactory.of(index_ref)
```

### Initial Retrieval

```python
# Create retrievers
bm25 = pt.BatchRetrieve(index, wmodel="BM25")
jm = pt.BatchRetrieve(index, wmodel="Hiemstra_LM")

# Queries for testing
queries = pd.DataFrame([
    {"qid": "1", "query": "revenge"},
    {"qid": "2", "query": "madness"},
    {"qid": "3", "query": "betrayal"}
])
```

```python
# Run baseline
bm25_results = bm25.transform(queries)
jm_results = jm.transform(queries)
```

### Pseudo-Relevance Feedback using RM3

```python
# Use RM3 with BM25 as the base retriever
rm3 = pt.rewrite.RM3(index, fb_terms=10, fb_docs=5) >> bm25
rm3_results = rm3.transform(queries)
```

### Evaluation

```python
# Example qrels (replace with actual data if available)
qrels = pd.DataFrame([
    {"qid": "1", "docno": "doc123", "label": 1},
    {"qid": "2", "docno": "doc456", "label": 1},
    {"qid": "3", "docno": "doc789", "label": 1}
])
```

```python
# Evaluation with MAP
pt.Utils.evaluate(bm25_results, qrels, metrics=["map"])
pt.Utils.evaluate(rm3_results, qrels, metrics=["map"])
```
